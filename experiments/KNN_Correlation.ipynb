### KNN Imputation
uses the K-Nearest Neighbors algorithm to impute missing values based on the similarity of rows.

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.impute import KNNImputer
```

```python
test_df_dropped.describe()

# data too large for KNN, hence select a subset from the dropnaed data
complete_cases = test_df_dropped.dropna()
sample = complete_cases.sample(n=10000, random_state=1)
```

```python
#use distance metrics to find the 'nearest neighbors'
scaler = MinMaxScaler()
test_df_dropped_scaled = scaler.fit_transform(sample.select_dtypes(include=[np.number]))
```

```python
#Use KNNImputer from scikit-learn to impute missing values
imputer = KNNImputer(n_neighbors=5)
test_df_imputed_scaled = imputer.fit_transform(test_df_dropped_scaled)
```

```python
test_df_imputed = scaler.inverse_transform(test_df_imputed_scaled) #After imputation, inverse transform the scaled data back to its original distribution.
sample[:] = test_df_imputed #Replace the old dataset with the imputed one for further analysis or modeling.
```

```python
missing_rate2=sample.apply(lambda x:sum(x.isnull())/len(x))
# print(missing_rate2.sort_values(ascending=False))
# plt.plot(missing_rate2.sort_values(ascending=False))
# reverse x,y of the plot, high resolution
fig=plt.figure(figsize=(15,15))
plt.plot(missing_rate2.sort_values().values,missing_rate2.sort_values().index)
plt.show()
```

```python
# now apply the method on the whole dataset the test_df_dropped
test_df_dropped_scaled = scaler.transform(test_df_dropped.select_dtypes(include=[np.number]))
test_df_imputed_scaled = imputer.transform(test_df_dropped_scaled) ## Apply the KNN imputer fitted on the sample to the full dataset
```

```python
test_df_filled = test_df_dropped.copy()
```

```python
test_df_imputed = scaler.inverse_transform(test_df_imputed_scaled) # Inverse scale the imputed data
test_df_dropped.loc[:, test_df_dropped.select_dtypes(include=[np.number]).columns] = test_df_imputed # Update the full dataset with the imputed values
```

```python
missing_rate2=test_df_dropped.apply(lambda x:sum(x.isnull())/len(x))

fig=plt.figure(figsize=(15,15))
plt.plot(missing_rate2.sort_values().values,missing_rate2.sort_values().index)
plt.show()
```
